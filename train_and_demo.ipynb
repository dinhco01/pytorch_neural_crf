```python
from transformers_predictor import TransformersNERPredictor

import nltk

nltk.download("punkt_tab")

sentences = [
    "Nền kinh tế Giao Châu từ thế kỷ VI - X bị tàn phá nghiêm trọng và không thể phát triển mạnh được do chiến tranh liên miên với Chiêm Thành, Chân Lạp, và Nam Chiếu, cùng với nạn cướp bóc, thuế má nặng nề, và thiên tai như hạn hán, lụt lội.",
    "Sau 17 năm biên soạn với sự tham gia của nhiều trí thức đương thời, bộ sách mang tên Tân Đường thư đã hoàn thành vào năm Gia Hựu thứ 5 (1060)",
]
sentences_tokenized = [nltk.word_tokenize(sent) for sent in sentences]

model_path = "model_files/phobert_base"
device = "cpu"
predictor = TransformersNERPredictor(model_path, cuda_device=device)
predictions = predictor.predict_and_display_with_entities(sentences_tokenized)
```
